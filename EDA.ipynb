**What is the shape of dataset after cleaning and transformation?**

```python
columns_to_check = ['HP', 'Price', 'Cylinders']
Q1 = df[columns_to_check].quantile(0.25)
Q3 = df[columns_to_check].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = ((df[columns_to_check] < lower_bound) | (df[columns_to_check] > upper_bound))
df = df[~outliers.any(axis=1)]
print(f"Shape after removing outliers: {df.shape}")
```

**Which car make appears the most frequently in the dataset?**

```python
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

**Which two features are most related?**

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
c = numeric_df.corr()
plt.figure(figsize=(10,5))
sns.heatmap(c, cmap="BrBG", annot=True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Transmission', y='MPG-H', data=df)
plt.title('Transmission vs Highway MPG')
plt.show()

plt.figure(figsize=(8, 6))
sns.violinplot(x='Drive Mode', y='Cylinders', data=df)
plt.title('Drive Mode vs Cylinders')
plt.show()
```
